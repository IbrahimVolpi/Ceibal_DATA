### Veamos cantidad de usuarios agrupados por cantidad de dias conectados en Montevideo en el 2020.

```python
%sql select dias_ingreso_crea, usuarios,t.total as total_de_usuarios, Round((sum(usuarios)/t.total)*100,2) as Porcentaje from days_log cross join (select sum(usuarios) as total from days_log) as t group by dias_ingreso_crea, usuarios, t.total ORDER BY dias_ingreso_crea ASC 
```

### Tomemos ahora los usuarios que ingresaron hasta 10 dias solamente, mismo año, mismo departamento 

```python
%sql select dias_ingreso_crea, usuarios,t.total as total_de_usuarios, Round((sum(usuarios)/t.total)*100,2) as Porcentaje from days_log cross join (select sum(usuarios) as total from days_log) as t group by dias_ingreso_crea, usuarios, t.total ORDER BY dias_ingreso_crea ASC LIMIT 11
```
